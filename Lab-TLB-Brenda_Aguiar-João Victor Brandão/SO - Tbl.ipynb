```python
import sys
import matplotlib.pyplot as plt
import numpy as np

# plt.plotfile(sys.argv[1],cols=(0,1),names=["Numero de Paginas","Tempo"],delimiter=" ")
# plt.show()
```

```python

def tblV(arq):
    arq = open(arq, "r")
    paginas = []
    tempo = []
    for i, linha in enumerate(arq):
        l = linha.split(" ")
        paginas.append(int(l[0]))
        tempo.append(float(l[1])*1000)
    return paginas, tempo
```

```python
paginas, tempo = tblV("tam_pagina.txt")

plt.figure(figsize=(10, 6))
plt.xlabel('Tempos', fontsize=15)
plt.ylabel('Paginas', fontsize=15)
plt.title('Tempo de acesso por pagina', fontsize=20)

plt.plot(paginas, tempo)
plt.yticks(np.arange(0, max(tempo)+0.002, step=0.004))
plt.show()
```

```python
paginas, tempo = tblV("tam_paginaOtimizado.txt")

plt.figure(figsize=(10, 6))
plt.xlabel('Tempos', fontsize=15)
plt.ylabel('Paginas', fontsize=15)
plt.title('Tempo de acesso por pagina', fontsize=20)

plt.plot(paginas, tempo)
plt.yticks(np.arange(0, max(tempo)+0.002, step=0.004))
plt.show()
```
